# UAL Clean

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 300)

ual_asm_df = pd.read_csv('Available Seat-Miles - All Airports - United.csv')
ual_flight_vol_df = pd.read_csv('Flights - All Airports - United.csv')
ual_fuel_df = pd.read_csv('Fuel - All Service - United.csv')
ual_load_factor_df = pd.read_csv('Load Factor - All Airports - United.csv')
ual_income_df = pd.read_csv('Net Income - All Regions - United.csv')
ual_operating_revenue_df = pd.read_csv('Operating Revenue - All Regions - United.csv')
ual_passenger_vol_df = pd.read_csv('Passengers - All Airports - United.csv')
ual_rpm_df = pd.read_csv('Revenue Passager-Miles - All Airports - United.csv')
ual_stock_df = pd.read_csv('UAL_stock_data.csv')

display(ual_asm_df.head())
display(ual_flight_vol_df.head())
display(ual_fuel_df.head())
display(ual_load_factor_df.head())
display(ual_income_df.head())
display(ual_operating_revenue_df.head())
display(ual_passenger_vol_df.head())
display(ual_rpm_df.head())
display(ual_stock_df.tail())
```

```python
# clean asm

# drop the total columns in order to convert to datetime. aggregation can easily replace total
ual_asm_df = ual_asm_df.loc[ual_asm_df['Month'] != "TOTAL"].copy()

# create datetime column to index compare to stock. 
ual_asm_df['date_str'] = ual_asm_df['Year'].astype(str) + '-' + ual_asm_df['Month'].astype(str) + '-01'
ual_asm_df['timestamp'] = pd.to_datetime(ual_asm_df['date_str'])

# put time stamp as first column for clarity
ual_asm_df = ual_asm_df[['timestamp'] + [col for col in ual_asm_df.columns if col != 'timestamp']]

# since these are monthly aggregates, set date to end of month
ual_asm_df['timestamp'] = ual_asm_df['timestamp'] + pd.offsets.MonthEnd(0)

#drop placeholder date_str
ual_asm_df.drop(columns=['date_str'], inplace=True)

# filter df for dates that align with stock data
start_date = pd.to_datetime('2016-10-01')
end_date = pd.to_datetime('2018-05-01')
ual_asm_df = ual_asm_df.loc[(ual_asm_df['timestamp'] >= start_date) & (ual_asm_df['timestamp'] < end_date)]

# convert dtypes
ual_asm_df['Domestic (Thousands)'] = ual_asm_df['Domestic (Thousands)'].str.replace(",", "").astype(float)
ual_asm_df['International (Thousands)'] = ual_asm_df['International (Thousands)'].str.replace(",", "").astype(float)
ual_asm_df['Total (Thousands)'] = ual_asm_df['Total (Thousands)'].str.replace(",", "").astype(float)
```

```python
# clean flight vol

# drop the total columns in order to convert to datetime. aggregation can easily replace total
ual_flight_vol_df = ual_flight_vol_df.loc[ual_flight_vol_df['Month'] != "TOTAL"].copy()

# create datetime column to index compare to stock. 
ual_flight_vol_df['date_str'] = ual_flight_vol_df['Year'].astype(str) + '-' + ual_flight_vol_df['Month'].astype(str) + '-01'
ual_flight_vol_df['timestamp'] = pd.to_datetime(ual_flight_vol_df['date_str'])

# put time stamp as first column for clarity
ual_flight_vol_df = ual_flight_vol_df[['timestamp'] + [col for col in ual_flight_vol_df.columns if col != 'timestamp']]

# since these are monthly aggregates, set date to end of month
ual_flight_vol_df['timestamp'] = ual_flight_vol_df['timestamp'] + pd.offsets.MonthEnd(0)

#drop placeholder date_str
ual_flight_vol_df.drop(columns=['date_str'], inplace=True)

# filter df for dates that align with stock data
start_date = pd.to_datetime('2016-10-01')
end_date = pd.to_datetime('2018-05-01')
ual_flight_vol_df = ual_flight_vol_df.loc[(ual_flight_vol_df['timestamp'] >= start_date) & (ual_flight_vol_df['timestamp'] < end_date)]

# clean dtypes
ual_flight_vol_df['DOMESTIC'] = ual_flight_vol_df['DOMESTIC'].str.replace(",", "").astype(float)
ual_flight_vol_df['INTERNATIONAL'] = ual_flight_vol_df['INTERNATIONAL'].str.replace(",", "").astype(float)
ual_flight_vol_df['TOTAL'] = ual_flight_vol_df['TOTAL'].str.replace(",", "").astype(float)
```

```python
# clean fuel df 

ual_fuel_df.dtypes

month_name_to_number = {
    'January': 1,
    'February': 2,
    'March': 3,
    'April': 4,
    'May': 5,
    'June': 6,
    'July': 7,
    'August': 8,
    'September': 9,
    'October': 10,
    'November': 11,
    'December': 12
}

ual_fuel_df['Month'] = ual_fuel_df['Month'].astype(str)
ual_fuel_df = ual_fuel_df.loc[ual_fuel_df['Month'] != "nan"].copy()
ual_fuel_df['Month'] = ual_fuel_df['Month'].map(month_name_to_number)

# create datetime column to index compare to stock. 
ual_fuel_df['date_str'] = ual_fuel_df['Year'].astype(str) + '-' + ual_fuel_df['Month'].astype(str) + '-01'
ual_fuel_df['timestamp'] = pd.to_datetime(ual_fuel_df['date_str'])

# put time stamp as first column for clarity
ual_fuel_df = ual_fuel_df[['timestamp'] + [col for col in ual_fuel_df.columns if col != 'timestamp']]

# since these are monthly aggregates, set date to end of month
ual_fuel_df['timestamp'] = ual_fuel_df['timestamp'] + pd.offsets.MonthEnd(0)

#drop placeholder date_str
ual_fuel_df.drop(columns=['date_str'], inplace=True)

# filter df for dates that align with stock data
start_date = pd.to_datetime('2016-10-01')
end_date = pd.to_datetime('2018-05-01')
ual_fuel_df = ual_fuel_df.loc[(ual_fuel_df['timestamp'] >= start_date) & (ual_fuel_df['timestamp'] < end_date)]

# clean dtypes
ual_fuel_df['Domestic Consumption\n(thousand gallons)'] = ual_fuel_df['Domestic Consumption\n(thousand gallons)'].str.replace(",", "").astype(float)
ual_fuel_df['Domestic Cost (thousand dollars)'] = ual_fuel_df['Domestic Cost (thousand dollars)'].str.replace(",", "").astype(float)
ual_fuel_df['Domestic Cost per Gallon (Dollars)'] = ual_fuel_df['Domestic Cost per Gallon (Dollars)'].str.replace(",", "").astype(float)
ual_fuel_df['International Consumption\n(thousand gallons)'] = ual_fuel_df['International Consumption\n(thousand gallons)'].str.replace(",", "").astype(float)
ual_fuel_df['International Cost (thousand dollars)'] = ual_fuel_df['International Cost (thousand dollars)'].str.replace(",", "").astype(float)
ual_fuel_df['International Cost per Gallon (Dollars)'] = ual_fuel_df['International Cost per Gallon (Dollars)'].str.replace(",", "").astype(float)
ual_fuel_df['Total Consumption\n(thousand gallons)'] = ual_fuel_df['Total Consumption\n(thousand gallons)'].str.replace(",", "").astype(float)
ual_fuel_df['Total Cost (thousand dollars)'] = ual_fuel_df['Total Cost (thousand dollars)'].str.replace(",", "").astype(float)
ual_fuel_df['Total Cost per Gallon (Dollars)'] = ual_fuel_df['Total Cost per Gallon (Dollars)'].str.replace(",", "").astype(float)
```

```python
# clean net income 

# dictionary to map quarters to their corresponding month and day
quarter_end_dates = {
    1: '03-31',
    2: '06-30',
    3: '09-30',
    4: '12-31'
}

# drop the total columns in order to convert to datetime. aggregation can easily replace total
ual_income_df = ual_income_df.loc[ual_income_df['Quarter'] != "TOTAL"].copy()

# convert quarter to numeric in order to use mapping 
ual_income_df['Quarter'] = pd.to_numeric(ual_income_df['Quarter'])

# create datetime column to index compare to stock. 
ual_income_df['date_str'] = ual_income_df['Year'].astype(str) + '-' + ual_income_df['Quarter'].map(quarter_end_dates)
ual_income_df['timestamp'] = pd.to_datetime(ual_income_df['date_str'])

# put time stamp as first column for clarity
ual_income_df = ual_income_df[['timestamp'] + [col for col in ual_income_df.columns if col != 'timestamp']]

# since these are monthly aggregates, set date to end of month
ual_income_df['timestamp'] = ual_income_df['timestamp'] + pd.offsets.MonthEnd(0)

#drop placeholder date_str
ual_income_df.drop(columns=['date_str'], inplace=True)

# filter df for dates that align with stock data
start_date = pd.to_datetime('2016-10-01')
end_date = pd.to_datetime('2018-05-01')
ual_income_df = ual_income_df.loc[(ual_income_df['timestamp'] >= start_date) & (ual_income_df['timestamp'] < end_date)]

# clean dtypes
ual_income_df['Domestic (Thousands)'] = ual_income_df['Domestic (Thousands)'].str.replace(",", "").astype(float)
ual_income_df['Latin America (Thousands)'] = ual_income_df['Latin America (Thousands)'].str.replace(",", "").astype(float)
ual_income_df['Total (Thousands)'] = ual_income_df['Total (Thousands)'].str.replace(",", "").astype(float)

# dtypes came clean
```
