# luv Clean

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 300)

luv_asm_df = pd.read_csv('Available Seat-Miles - All Airports - Southwest.csv')
luv_flight_vol_df = pd.read_csv('Flights - All Airports - Southwest.csv')
luv_fuel_df = pd.read_csv('Fuel - All Service - Southwest.csv')
luv_load_factor_df = pd.read_csv('Load Factor - All Airports - Southwest.csv')
luv_income_df = pd.read_csv('Net Income - All Regions - Southwest.csv')
luv_operating_revenue_df = pd.read_csv('Operating Revenue - All Regions - Southwest.csv')
luv_passenger_vol_df = pd.read_csv('Passengers - All Airports - Southwest.csv')
luv_rpm_df = pd.read_csv('Revenue Passager-Miles - All Airports - Southwest.csv')
luv_stock_df = pd.read_csv('LUV_stock_data.csv')
luv_baggage_df = pd.read_csv('Baggage Fees - All Airports - Southwest.csv')

display(luv_asm_df.head())
display(luv_flight_vol_df.head())
display(luv_fuel_df.head())
display(luv_load_factor_df.head())
display(luv_income_df.head())
display(luv_operating_revenue_df.head())
display(luv_passenger_vol_df.head())
display(luv_rpm_df.head())
display(luv_stock_df.tail())
```

```python
# clean baggage

# dictionary to map quarters to their corresponding month and day
quarter_end_dates = {
    1: '03-31',
    2: '06-30',
    3: '09-30',
    4: '12-31'
}

# drop the total columns in order to convert to datetime. aggregation can easily replace total
luv_baggage_df = luv_baggage_df.loc[luv_baggage_df['Quarter'] != "TOTAL"].copy()

# convert quarter to numeric in order to use mapping 
luv_baggage_df['Quarter'] = pd.to_numeric(luv_baggage_df['Quarter'])

# create datetime column to index compare to stock. 
luv_baggage_df['date_str'] = luv_baggage_df['Year'].astype(str) + '-' + luv_baggage_df['Quarter'].map(quarter_end_dates)
luv_baggage_df['timestamp'] = pd.to_datetime(luv_baggage_df['date_str'])

# put time stamp as first column for clarity
luv_baggage_df = luv_baggage_df[['timestamp'] + [col for col in luv_baggage_df.columns if col != 'timestamp']]

# since these are monthly aggregates, set date to end of month
luv_baggage_df['timestamp'] = luv_baggage_df['timestamp'] + pd.offsets.MonthEnd(0)

#drop placeholder date_str
luv_baggage_df.drop(columns=['date_str'], inplace=True)

# filter df for dates that align with stock data
start_date = pd.to_datetime('2016-10-01')
end_date = pd.to_datetime('2018-05-01')
luv_baggage_df = luv_baggage_df.loc[(luv_baggage_df['timestamp'] >= start_date) & (luv_baggage_df['timestamp'] < end_date)]

# clean dtypes
luv_baggage_df['Total (Thousands)'] = luv_baggage_df['Total (Thousands)'].str.replace(",", "").astype(float)
```

```python
# clean fuel df 

luv_fuel_df.dtypes

month_name_to_number = {
    'January': 1,
    'February': 2,
    'March': 3,
    'April': 4,
    'May': 5,
    'June': 6,
    'July': 7,
    'August': 8,
    'September': 9,
    'October': 10,
    'November': 11,
    'December': 12
}

luv_fuel_df['Month'] = luv_fuel_df['Month'].astype(str)
luv_fuel_df = luv_fuel_df.loc[luv_fuel_df['Month'] != "nan"].copy()
luv_fuel_df['Month'] = luv_fuel_df['Month'].map(month_name_to_number)

# create datetime column to index compare to stock. 
luv_fuel_df['date_str'] = luv_fuel_df['Year'].astype(str) + '-' + luv_fuel_df['Month'].astype(str) + '-01'
luv_fuel_df['timestamp'] = pd.to_datetime(luv_fuel_df['date_str'])

# put time stamp as first column for clarity
luv_fuel_df = luv_fuel_df[['timestamp'] + [col for col in luv_fuel_df.columns if col != 'timestamp']]

# since these are monthly aggregates, set date to end of month
luv_fuel_df['timestamp'] = luv_fuel_df['timestamp'] + pd.offsets.MonthEnd(0)

#drop placeholder date_str
luv_fuel_df.drop(columns=['date_str'], inplace=True)

# filter df for dates that align with stock data
start_date = pd.to_datetime('2016-10-01')
end_date = pd.to_datetime('2018-05-01')
luv_fuel_df = luv_fuel_df.loc[(luv_fuel_df['timestamp'] >= start_date) & (luv_fuel_df['timestamp'] < end_date)]

# clean dtypes
luv_fuel_df['Domestic Consumption\n(thousand gallons)'] = luv_fuel_df['Domestic Consumption\n(thousand gallons)'].str.replace(",", "").astype(float)
luv_fuel_df['Domestic Cost (thousand dollars)'] = luv_fuel_df['Domestic Cost (thousand dollars)'].str.replace(",", "").astype(float)
luv_fuel_df['Domestic Cost per Gallon (Dollars)'] = luv_fuel_df['Domestic Cost per Gallon (Dollars)'].str.replace(",", "").astype(float)
luv_fuel_df['International Consumption\n(thousand gallons)'] = luv_fuel_df['International Consumption\n(thousand gallons)'].str.replace(",", "").astype(float)
luv_fuel_df['International Cost (thousand dollars)'] = luv_fuel_df['International Cost (thousand dollars)'].str.replace(",", "").astype(float)
luv_fuel_df['International Cost per Gallon (Dollars)'] = luv_fuel_df['International Cost per Gallon (Dollars)'].str.replace(",", "").astype(float)
luv_fuel_df['Total Consumption\n(thousand gallons)'] = luv_fuel_df['Total Consumption\n(thousand gallons)'].str.replace(",", "").astype(float)
luv_fuel_df['Total Cost (thousand dollars)'] = luv_fuel_df['Total Cost (thousand dollars)'].str.replace(",", "").astype(float)
luv_fuel_df['Total Cost per Gallon (Dollars)'] = luv_fuel_df['Total Cost per Gallon (Dollars)'].str.replace(",", "").astype(float)
```
