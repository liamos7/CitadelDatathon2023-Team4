# DAL Clean

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 300)

dal_asm_df = pd.read_csv('Available Seat-Miles - All Airports - Delta.csv')
dal_flight_vol_df = pd.read_csv('Flights - All Airports - Delta.csv')
dal_fuel_df = pd.read_csv('Fuel - All Service - Delta.csv')
dal_load_factor_df = pd.read_csv('Load Factor - All Airports - Delta.csv')
dal_income_df = pd.read_csv('Net Income - All Regions - Delta.csv')
dal_operating_revenue_df = pd.read_csv('Operating Revenue - All Regions - Delta.csv')
dal_passenger_vol_df = pd.read_csv('Passengers - All Airports - Delta.csv')
dal_rpm_df = pd.read_csv('Revenue Passager-Miles - All Airports - Delta.csv')
dal_stock_df = pd.read_csv('DAL_stock_data.csv')
dal_baggage_df = pd.read_csv('Baggage Fees - All Airports - Delta.csv')

display(dal_asm_df.head())
display(dal_flight_vol_df.head())
display(dal_fuel_df.head())
display(dal_load_factor_df.head())
display(dal_income_df.head())
display(dal_operating_revenue_df.head())
display(dal_passenger_vol_df.head())
display(dal_rpm_df.head())
display(dal_stock_df.tail())
```

```python
# clean baggage

# dictionary to map quarters to their corresponding month and day
quarter_end_dates = {
    1: '03-31',
    2: '06-30',
    3: '09-30',
    4: '12-31'
}

# drop the total columns in order to convert to datetime. aggregation can easily replace total
dal_baggage_df = dal_baggage_df.loc[dal_baggage_df['Quarter'] != "TOTAL"].copy()

# convert quarter to numeric in order to use mapping 
dal_baggage_df['Quarter'] = pd.to_numeric(dal_baggage_df['Quarter'])

# create datetime column to index compare to stock. 
dal_baggage_df['date_str'] = dal_baggage_df['Year'].astype(str) + '-' + dal_baggage_df['Quarter'].map(quarter_end_dates)
dal_baggage_df['timestamp'] = pd.to_datetime(dal_baggage_df['date_str'])

# put time stamp as first column for clarity
dal_baggage_df = dal_baggage_df[['timestamp'] + [col for col in dal_baggage_df.columns if col != 'timestamp']]

# since these are monthly aggregates, set date to end of month
dal_baggage_df['timestamp'] = dal_baggage_df['timestamp'] + pd.offsets.MonthEnd(0)

#drop placeholder date_str
dal_baggage_df.drop(columns=['date_str'], inplace=True)

# filter df for dates that align with stock data
start_date = pd.to_datetime('2016-10-01')
end_date = pd.to_datetime('2018-05-01')
dal_baggage_df = dal_baggage_df.loc[(dal_baggage_df['timestamp'] >= start_date) & (dal_baggage_df['timestamp'] < end_date)]

# clean dtypes
dal_baggage_df['Total (Thousands)'] = dal_baggage_df['Total (Thousands)'].str.replace(",", "").astype(float)
```
