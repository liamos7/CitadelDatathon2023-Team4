# HA Clean

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 300)

ha_asm_df = pd.read_csv('Available Seat-Miles - All Airports - Hawaiian.csv')
ha_flight_vol_df = pd.read_csv('Flights - All Airports - Hawaiian.csv')
ha_fuel_df = pd.read_csv('Fuel - All Service - Hawaiian.csv')
ha_load_factor_df = pd.read_csv('Load Factor - All Airports - Hawaiian.csv')
ha_income_df = pd.read_csv('Net Income - All Regions - Hawaiian.csv')
ha_operating_revenue_df = pd.read_csv('Operating Revenue - All Regions - Hawaiian.csv')
ha_passenger_vol_df = pd.read_csv('Passengers - All Airports - Hawaiian.csv')
ha_rpm_df = pd.read_csv('Revenue Passager-Miles - All Airports - Hawaiian.csv')
ha_stock_df = pd.read_csv('HA_stock_data.csv')
ha_baggage_df = pd.read_csv('Baggage Fees - All Airports - Hawaiian.csv')

display(ha_asm_df.head())
display(ha_flight_vol_df.head())
display(ha_fuel_df.head())
display(ha_load_factor_df.head())
display(ha_income_df.head())
display(ha_operating_revenue_df.head())
display(ha_passenger_vol_df.head())
display(ha_rpm_df.head())
display(ha_stock_df.tail())
```

```python
# clean baggage

# dictionary to map quarters to their corresponding month and day
quarter_end_dates = {
    1: '03-31',
    2: '06-30',
    3: '09-30',
    4: '12-31'
}

# drop the total columns in order to convert to datetime. aggregation can easily replace total
ha_baggage_df = ha_baggage_df.loc[ha_baggage_df['Quarter'] != "TOTAL"].copy()

# convert quarter to numeric in order to use mapping 
ha_baggage_df['Quarter'] = pd.to_numeric(ha_baggage_df['Quarter'])

# create datetime column to index compare to stock. 
ha_baggage_df['date_str'] = ha_baggage_df['Year'].astype(str) + '-' + ha_baggage_df['Quarter'].map(quarter_end_dates)
ha_baggage_df['timestamp'] = pd.to_datetime(ha_baggage_df['date_str'])

# put time stamp as first column for clarity
ha_baggage_df = ha_baggage_df[['timestamp'] + [col for col in ha_baggage_df.columns if col != 'timestamp']]

# since these are monthly aggregates, set date to end of month
ha_baggage_df['timestamp'] = ha_baggage_df['timestamp'] + pd.offsets.MonthEnd(0)

#drop placeholder date_str
ha_baggage_df.drop(columns=['date_str'], inplace=True)

# filter df for dates that align with stock data
start_date = pd.to_datetime('2016-10-01')
end_date = pd.to_datetime('2018-05-01')
ha_baggage_df = ha_baggage_df.loc[(ha_baggage_df['timestamp'] >= start_date) & (ha_baggage_df['timestamp'] < end_date)]

# clean dtypes
ha_baggage_df['Total (Thousands)'] = ha_baggage_df['Total (Thousands)'].str.replace(",", "").astype(float)
```

```python
# clean fuel df 

ha_fuel_df.dtypes

month_name_to_number = {
    'January': 1,
    'February': 2,
    'March': 3,
    'April': 4,
    'May': 5,
    'June': 6,
    'July': 7,
    'August': 8,
    'September': 9,
    'October': 10,
    'November': 11,
    'December': 12
}

ha_fuel_df['Month'] = ha_fuel_df['Month'].astype(str)
ha_fuel_df = ha_fuel_df.loc[ha_fuel_df['Month'] != "nan"].copy()
ha_fuel_df['Month'] = ha_fuel_df['Month'].map(month_name_to_number)

# create datetime column to index compare to stock. 
ha_fuel_df['date_str'] = ha_fuel_df['Year'].astype(str) + '-' + ha_fuel_df['Month'].astype(str) + '-01'
ha_fuel_df['timestamp'] = pd.to_datetime(ha_fuel_df['date_str'])

# put time stamp as first column for clarity
ha_fuel_df = ha_fuel_df[['timestamp'] + [col for col in ha_fuel_df.columns if col != 'timestamp']]

# since these are monthly aggregates, set date to end of month
ha_fuel_df['timestamp'] = ha_fuel_df['timestamp'] + pd.offsets.MonthEnd(0)

#drop placeholder date_str
ha_fuel_df.drop(columns=['date_str'], inplace=True)

# filter df for dates that align with stock data
start_date = pd.to_datetime('2016-10-01')
end_date = pd.to_datetime('2018-05-01')
ha_fuel_df = ha_fuel_df.loc[(ha_fuel_df['timestamp'] >= start_date) & (ha_fuel_df['timestamp'] < end_date)]

# clean dtypes
ha_fuel_df['Domestic Consumption\n(thousand gallons)'] = ha_fuel_df['Domestic Consumption\n(thousand gallons)'].str.replace(",", "").astype(float)
ha_fuel_df['Domestic Cost (thousand dollars)'] = ha_fuel_df['Domestic Cost (thousand dollars)'].str.replace(",", "").astype(float)
ha_fuel_df['Domestic Cost per Gallon (Dollars)'] = ha_fuel_df['Domestic Cost per Gallon (Dollars)'].str.replace(",", "").astype(float)
ha_fuel_df['International Consumption\n(thousand gallons)'] = ha_fuel_df['International Consumption\n(thousand gallons)'].str.replace(",", "").astype(float)
ha_fuel_df['International Cost (thousand dollars)'] = ha_fuel_df['International Cost (thousand dollars)'].str.replace(",", "").astype(float)
ha_fuel_df['International Cost per Gallon (Dollars)'] = ha_fuel_df['International Cost per Gallon (Dollars)'].str.replace(",", "").astype(float)
ha_fuel_df['Total Consumption\n(thousand gallons)'] = ha_fuel_df['Total Consumption\n(thousand gallons)'].str.replace(",", "").astype(float)
ha_fuel_df['Total Cost (thousand dollars)'] = ha_fuel_df['Total Cost (thousand dollars)'].str.replace(",", "").astype(float)
ha_fuel_df['Total Cost per Gallon (Dollars)'] = ha_fuel_df['Total Cost per Gallon (Dollars)'].str.replace(",", "").astype(float)
```
