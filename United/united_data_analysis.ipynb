```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 300)

ual_asm_df = pd.read_csv('Available Seat-Miles - All Airports - United.csv')
ual_flight_vol_df = pd.read_csv('Flights - All Airports - United.csv')
ual_fuel_df = pd.read_csv('Fuel - All Service - United.csv')
ual_load_factor_df = pd.read_csv('Load Factor - All Airports - United.csv')
ual_income_df = pd.read_csv('Net Income - All Regions - United.csv')
ual_operating_revenue_df = pd.read_csv('Operating Revenue - All Regions - United.csv')
ual_passenger_vol_df = pd.read_csv('Passengers - All Airports - United.csv')
ual_rpm_df = pd.read_csv('Revenue Passager-Miles - All Airports - United.csv')
ual_stock_df = pd.read_csv('UAL_stock_data.csv')

display(ual_asm_df.head())
display(ual_flight_vol_df.head())
display(ual_fuel_df.head())
display(ual_load_factor_df.head())
display(ual_income_df.head())
display(ual_operating_revenue_df.head())
display(ual_passenger_vol_df.head())
display(ual_rpm_df.head())
display(ual_stock_df.tail())
```

```python
# clean asm

# drop the total columns in order to convert to datetime. aggregation can easily replace total
ual_asm_df = ual_asm_df.loc[ual_asm_df['Month'] != "TOTAL"].copy()

# create datetime column to index compare to stock. 
ual_asm_df['date_str'] = ual_asm_df['Year'].astype(str) + '-' + ual_asm_df['Month'].astype(str) + '-01'
ual_asm_df['timestamp'] = pd.to_datetime(ual_asm_df['date_str'])

# put time stamp as first column for clarity
ual_asm_df = ual_asm_df[['timestamp'] + [col for col in ual_asm_df.columns if col != 'timestamp']]

# since these are monthly aggregates, set date to end of month
ual_asm_df['timestamp'] = ual_asm_df['timestamp'] + pd.offsets.MonthEnd(0)

#drop placeholder date_str
ual_asm_df.drop(columns=['date_str'], inplace=True)

# filter df for dates that align with stock data
start_date = pd.to_datetime('2016-10-01')
end_date = pd.to_datetime('2018-05-01')
ual_asm_df = ual_asm_df.loc[(ual_asm_df['timestamp'] >= start_date) & (ual_asm_df['timestamp'] < end_date)]

```

```python
# clean net income 

# dictionary to map quarters to their corresponding month and day
quarter_end_dates = {
    1: '03-31',
    2: '06-30',
    3: '09-30',
    4: '12-31'
}

# drop the total columns in order to convert to datetime. aggregation can easily replace total
ual_income_df = ual_income_df.loc[ual_income_df['Quarter'] != "TOTAL"].copy()

# convert quarter to numeric in order to use mapping 
ual_income_df['Quarter'] = pd.to_numeric(ual_income_df['Quarter'])

# create datetime column to index compare to stock. 
ual_income_df['date_str'] = ual_income_df['Year'].astype(str) + '-' + ual_income_df['Quarter'].map(quarter_end_dates)
ual_income_df['timestamp'] = pd.to_datetime(ual_income_df['date_str'])

# put time stamp as first column for clarity
ual_income_df = ual_income_df[['timestamp'] + [col for col in ual_income_df.columns if col != 'timestamp']]

# since these are monthly aggregates, set date to end of month
ual_income_df['timestamp'] = ual_income_df['timestamp'] + pd.offsets.MonthEnd(0)

#drop placeholder date_str
ual_income_df.drop(columns=['date_str'], inplace=True)

# filter df for dates that align with stock data
start_date = pd.to_datetime('2016-10-01')
end_date = pd.to_datetime('2018-05-01')
ual_income_df = ual_income_df.loc[(ual_income_df['timestamp'] >= start_date) & (ual_income_df['timestamp'] < end_date)]
```
