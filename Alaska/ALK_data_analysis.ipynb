# ALK Clean

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 300)

alk_asm_df = pd.read_csv('Available Seat-Miles - All Airports - Alaska.csv')
alk_flight_vol_df = pd.read_csv('Flights - All Airports - Alaska.csv')
alk_fuel_df = pd.read_csv('Fuel - All Service - Alaska.csv')
alk_load_factor_df = pd.read_csv('Load Factor - All Airports - Alaska.csv')
alk_income_df = pd.read_csv('Net Income - All Regions - Alaska.csv')
alk_operating_revenue_df = pd.read_csv('Operating Revenue - All Regions - Alaska.csv')
alk_passenger_vol_df = pd.read_csv('Passengers - All Airports - Alaska.csv')
alk_rpm_df = pd.read_csv('Revenue Passager-Miles - All Airports - Alaska.csv')
alk_stock_df = pd.read_csv('ALK_stock_data.csv')
alk_baggage_df = pd.read_csv('Baggage Fees - All Airports - Alaska.csv')

display(alk_asm_df.head())
display(alk_flight_vol_df.head())
display(alk_fuel_df.head())
display(alk_load_factor_df.head())
display(alk_income_df.head())
display(alk_operating_revenue_df.head())
display(alk_passenger_vol_df.head())
display(alk_rpm_df.head())
display(alk_stock_df.tail())
```

```python
# clean baggage

# dictionary to map quarters to their corresponding month and day
quarter_end_dates = {
    1: '03-31',
    2: '06-30',
    3: '09-30',
    4: '12-31'
}

# drop the total columns in order to convert to datetime. aggregation can easily replace total
alk_baggage_df = alk_baggage_df.loc[alk_baggage_df['Quarter'] != "TOTAL"].copy()

# convert quarter to numeric in order to use mapping 
alk_baggage_df['Quarter'] = pd.to_numeric(alk_baggage_df['Quarter'])

# create datetime column to index compare to stock. 
alk_baggage_df['date_str'] = alk_baggage_df['Year'].astype(str) + '-' + alk_baggage_df['Quarter'].map(quarter_end_dates)
alk_baggage_df['timestamp'] = pd.to_datetime(alk_baggage_df['date_str'])

# put time stamp as first column for clarity
alk_baggage_df = alk_baggage_df[['timestamp'] + [col for col in alk_baggage_df.columns if col != 'timestamp']]

# since these are monthly aggregates, set date to end of month
alk_baggage_df['timestamp'] = alk_baggage_df['timestamp'] + pd.offsets.MonthEnd(0)

#drop placeholder date_str
alk_baggage_df.drop(columns=['date_str'], inplace=True)

# filter df for dates that align with stock data
start_date = pd.to_datetime('2016-10-01')
end_date = pd.to_datetime('2018-05-01')
alk_baggage_df = alk_baggage_df.loc[(alk_baggage_df['timestamp'] >= start_date) & (alk_baggage_df['timestamp'] < end_date)]

# clean dtypes
alk_baggage_df['Total (Thousands)'] = alk_baggage_df['Total (Thousands)'].str.replace(",", "").astype(float)
```

```python
# clean asm

# drop the total columns in order to convert to datetime. aggregation can easily replace total
alk_asm_df = alk_asm_df.loc[alk_asm_df['Month'] != "TOTAL"].copy()

# create datetime column to index compare to stock. 
alk_asm_df['date_str'] = alk_asm_df['Year'].astype(str) + '-' + alk_asm_df['Month'].astype(str) + '-01'
alk_asm_df['timestamp'] = pd.to_datetime(alk_asm_df['date_str'])

# put time stamp as first column for clarity
alk_asm_df = alk_asm_df[['timestamp'] + [col for col in alk_asm_df.columns if col != 'timestamp']]

# since these are monthly aggregates, set date to end of month
alk_asm_df['timestamp'] = alk_asm_df['timestamp'] + pd.offsets.MonthEnd(0)

#drop placeholder date_str
alk_asm_df.drop(columns=['date_str'], inplace=True)

# filter df for dates that align with stock data
start_date = pd.to_datetime('2016-10-01')
end_date = pd.to_datetime('2018-05-01')
alk_asm_df = alk_asm_df.loc[(alk_asm_df['timestamp'] >= start_date) & (alk_asm_df['timestamp'] < end_date)]

# convert dtypes
alk_asm_df['Domestic (Thousands)'] = alk_asm_df['Domestic (Thousands)'].str.replace(",", "").astype(float)
alk_asm_df['International (Thousands)'] = alk_asm_df['International (Thousands)'].str.replace(",", "").astype(float)
alk_asm_df['Total (Thousands)'] = alk_asm_df['Total (Thousands)'].str.replace(",", "").astype(float)
```

```python
# clean fuel df 

alk_fuel_df.dtypes

month_name_to_number = {
    'January': 1,
    'February': 2,
    'March': 3,
    'April': 4,
    'May': 5,
    'June': 6,
    'July': 7,
    'August': 8,
    'September': 9,
    'October': 10,
    'November': 11,
    'December': 12
}

alk_fuel_df['Month'] = alk_fuel_df['Month'].astype(str)
alk_fuel_df = alk_fuel_df.loc[alk_fuel_df['Month'] != "nan"].copy()
alk_fuel_df['Month'] = alk_fuel_df['Month'].map(month_name_to_number)

# create datetime column to index compare to stock. 
alk_fuel_df['date_str'] = alk_fuel_df['Year'].astype(str) + '-' + alk_fuel_df['Month'].astype(str) + '-01'
alk_fuel_df['timestamp'] = pd.to_datetime(alk_fuel_df['date_str'])

# put time stamp as first column for clarity
alk_fuel_df = alk_fuel_df[['timestamp'] + [col for col in alk_fuel_df.columns if col != 'timestamp']]

# since these are monthly aggregates, set date to end of month
alk_fuel_df['timestamp'] = alk_fuel_df['timestamp'] + pd.offsets.MonthEnd(0)

#drop placeholder date_str
alk_fuel_df.drop(columns=['date_str'], inplace=True)

# filter df for dates that align with stock data
start_date = pd.to_datetime('2016-10-01')
end_date = pd.to_datetime('2018-05-01')
alk_fuel_df = alk_fuel_df.loc[(alk_fuel_df['timestamp'] >= start_date) & (alk_fuel_df['timestamp'] < end_date)]

# clean dtypes
alk_fuel_df['Domestic Consumption\n(thousand gallons)'] = alk_fuel_df['Domestic Consumption\n(thousand gallons)'].str.replace(",", "").astype(float)
alk_fuel_df['Domestic Cost (thousand dollars)'] = alk_fuel_df['Domestic Cost (thousand dollars)'].str.replace(",", "").astype(float)
alk_fuel_df['Domestic Cost per Gallon (Dollars)'] = alk_fuel_df['Domestic Cost per Gallon (Dollars)'].str.replace(",", "").astype(float)
alk_fuel_df['International Consumption\n(thousand gallons)'] = alk_fuel_df['International Consumption\n(thousand gallons)'].str.replace(",", "").astype(float)
alk_fuel_df['International Cost (thousand dollars)'] = alk_fuel_df['International Cost (thousand dollars)'].str.replace(",", "").astype(float)
alk_fuel_df['International Cost per Gallon (Dollars)'] = alk_fuel_df['International Cost per Gallon (Dollars)'].str.replace(",", "").astype(float)
alk_fuel_df['Total Consumption\n(thousand gallons)'] = alk_fuel_df['Total Consumption\n(thousand gallons)'].str.replace(",", "").astype(float)
alk_fuel_df['Total Cost (thousand dollars)'] = alk_fuel_df['Total Cost (thousand dollars)'].str.replace(",", "").astype(float)
alk_fuel_df['Total Cost per Gallon (Dollars)'] = alk_fuel_df['Total Cost per Gallon (Dollars)'].str.replace(",", "").astype(float)
```
