# JBLU Clean

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 300)

jblu_asm_df = pd.read_csv('Available Seat-Miles - All Airports - JetBlue.csv')
jblu_flight_vol_df = pd.read_csv('Flights - All Airports - JetBlue.csv')
jblu_fuel_df = pd.read_csv('Fuel - All Service - JetBlue.csv')
jblu_load_factor_df = pd.read_csv('Load Factor - All Airports - JetBlue.csv')
jblu_income_df = pd.read_csv('Net Income - All Regions - JetBlue.csv')
jblu_operating_revenue_df = pd.read_csv('Operating Revenue - All Regions - JetBlue.csv')
jblu_passenger_vol_df = pd.read_csv('Passengers - All Airports - JetBlue.csv')
jblu_rpm_df = pd.read_csv('Revenue Passager-Miles - All Airports - JetBlue.csv')
jblu_stock_df = pd.read_csv('JBLU_stock_data.csv')
jblu_baggage_df = pd.read_csv('Baggage Fees - All Airports - JetBlue.csv')

display(jblu_asm_df.head())
display(jblu_flight_vol_df.head())
display(jblu_fuel_df.head())
display(jblu_load_factor_df.head())
display(jblu_income_df.head())
display(jblu_operating_revenue_df.head())
display(jblu_passenger_vol_df.head())
display(jblu_rpm_df.head())
display(jblu_stock_df.tail())
```

```python
# clean baggage

# dictionary to map quarters to their corresponding month and day
quarter_end_dates = {
    1: '03-31',
    2: '06-30',
    3: '09-30',
    4: '12-31'
}

# drop the total columns in order to convert to datetime. aggregation can easily replace total
jblu_baggage_df = jblu_baggage_df.loc[jblu_baggage_df['Quarter'] != "TOTAL"].copy()

# convert quarter to numeric in order to use mapping 
jblu_baggage_df['Quarter'] = pd.to_numeric(jblu_baggage_df['Quarter'])

# create datetime column to index compare to stock. 
jblu_baggage_df['date_str'] = jblu_baggage_df['Year'].astype(str) + '-' + jblu_baggage_df['Quarter'].map(quarter_end_dates)
jblu_baggage_df['timestamp'] = pd.to_datetime(jblu_baggage_df['date_str'])

# put time stamp as first column for clarity
jblu_baggage_df = jblu_baggage_df[['timestamp'] + [col for col in jblu_baggage_df.columns if col != 'timestamp']]

# since these are monthly aggregates, set date to end of month
jblu_baggage_df['timestamp'] = jblu_baggage_df['timestamp'] + pd.offsets.MonthEnd(0)

#drop placeholder date_str
jblu_baggage_df.drop(columns=['date_str'], inplace=True)

# filter df for dates that align with stock data
start_date = pd.to_datetime('2016-10-01')
end_date = pd.to_datetime('2018-05-01')
jblu_baggage_df = jblu_baggage_df.loc[(jblu_baggage_df['timestamp'] >= start_date) & (jblu_baggage_df['timestamp'] < end_date)]

# clean dtypes
jblu_baggage_df['Total (Thousands)'] = jblu_baggage_df['Total (Thousands)'].str.replace(",", "").astype(float)
```
