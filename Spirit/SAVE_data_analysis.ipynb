```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 300)

save_asm_df = pd.read_csv('Available Seat-Miles - All Airports - Spirit.csv')
save_flight_vol_df = pd.read_csv('Flights - All Airports - Spirit.csv')
save_fuel_df = pd.read_csv('Fuel - All Service - Spirit.csv')
save_load_factor_df = pd.read_csv('Load Factor - All Airports - Spirit.csv')
save_income_df = pd.read_csv('Net Income - All Regions - Spirit.csv')
save_operating_revenue_df = pd.read_csv('Operating Revenue - All Regions - Spirit.csv')
save_passenger_vol_df = pd.read_csv('Passengers - All Airports - Spirit.csv')
save_rpm_df = pd.read_csv('Revenue Passager-Miles - All Airports - Spirit.csv')
save_stock_df = pd.read_csv('SAVE_stock_data.csv')
save_baggage_df = pd.read_csv('Baggage Fees - All Airports - Spirit.csv')

display(save_asm_df.head())
display(save_flight_vol_df.head())
display(save_fuel_df.head())
display(save_load_factor_df.head())
display(save_income_df.head())
display(save_operating_revenue_df.head())
display(save_passenger_vol_df.head())
display(save_rpm_df.head())
display(save_stock_df.tail())
```

```python
# clean baggage

# dictionary to map quarters to their corresponding month and day
quarter_end_dates = {
    1: '03-31',
    2: '06-30',
    3: '09-30',
    4: '12-31'
}

# drop the total columns in order to convert to datetime. aggregation can easily replace total
save_baggage_df = save_baggage_df.loc[save_baggage_df['Quarter'] != "TOTAL"].copy()

# convert quarter to numeric in order to use mapping 
save_baggage_df['Quarter'] = pd.to_numeric(save_baggage_df['Quarter'])

# create datetime column to index compare to stock. 
save_baggage_df['date_str'] = save_baggage_df['Year'].astype(str) + '-' + save_baggage_df['Quarter'].map(quarter_end_dates)
save_baggage_df['timestamp'] = pd.to_datetime(save_baggage_df['date_str'])

# put time stamp as first column for clarity
save_baggage_df = save_baggage_df[['timestamp'] + [col for col in save_baggage_df.columns if col != 'timestamp']]

# since these are monthly aggregates, set date to end of month
save_baggage_df['timestamp'] = save_baggage_df['timestamp'] + pd.offsets.MonthEnd(0)

#drop placeholder date_str
save_baggage_df.drop(columns=['date_str'], inplace=True)

# filter df for dates that align with stock data
start_date = pd.to_datetime('2016-10-01')
end_date = pd.to_datetime('2018-05-01')
save_baggage_df = save_baggage_df.loc[(save_baggage_df['timestamp'] >= start_date) & (save_baggage_df['timestamp'] < end_date)]

# clean dtypes
save_baggage_df['Total (Thousands)'] = save_baggage_df['Total (Thousands)'].str.replace(",", "").astype(float)
```
