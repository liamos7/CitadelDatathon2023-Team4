```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', 300)

aal_asm_df = pd.read_csv('Available Seat-Miles - All Airports - American.csv')
aal_flight_vol_df = pd.read_csv('Flights - All Airports - American.csv')
aal_fuel_df = pd.read_csv('Fuel - All Service - American.csv')
aal_load_factor_df = pd.read_csv('Load Factor - All Airports - American.csv')
aal_income_df = pd.read_csv('Net Income - All Regions - American.csv')
aal_operating_revenue_df = pd.read_csv('Operating Revenue - All Regions - American.csv')
aal_passenger_vol_df = pd.read_csv('Passengers - All Airports - American.csv')
aal_rpm_df = pd.read_csv('Revenue Passager-Miles - All Airports - American.csv')
aal_stock_df = pd.read_csv('AAL_stock_data.csv')
aal_baggage_df = pd.read_csv('Baggage Fees - All Airports - American.csv')

display(aal_asm_df.head())
display(aal_flight_vol_df.head())
display(aal_fuel_df.head())
display(aal_load_factor_df.head())
display(aal_income_df.head())
display(aal_operating_revenue_df.head())
display(aal_passenger_vol_df.head())
display(aal_rpm_df.head())
display(aal_stock_df.tail())
```

```python
# clean baggage

# dictionary to map quarters to their corresponding month and day
quarter_end_dates = {
    1: '03-31',
    2: '06-30',
    3: '09-30',
    4: '12-31'
}

# drop the total columns in order to convert to datetime. aggregation can easily replace total
aal_baggage_df = aal_baggage_df.loc[aal_baggage_df['Quarter'] != "TOTAL"].copy()

# convert quarter to numeric in order to use mapping 
aal_baggage_df['Quarter'] = pd.to_numeric(aal_baggage_df['Quarter'])

# convert to quarters, then resample to months
aal_baggage_df['date_str'] = aal_baggage_df['Year'].astype(str) + '-' + aal_baggage_df['Quarter'].map(quarter_end_dates)
aal_baggage_df['timestamp'] = pd.PeriodIndex(aal_baggage_df['date_str'], freq='Q')
aal_baggage_df.set_index('timestamp', inplace=True)
aal_baggage_df = aal_baggage_df.resample('M').ffill()


#drop placeholder date_str
aal_baggage_df.drop(columns=['date_str'], inplace=True)

# select the rows with dates corresponding to other data (October 2016-April 2018)
aal_baggage_df = aal_baggage_df.reset_index()
start_date = pd.Period('2016-10-01', freq = 'M')
end_date = pd.Period('2018-05-01', freq='M')
aal_baggage_df = aal_baggage_df.loc[(aal_baggage_df['timestamp'] >= start_date) & (aal_baggage_df['timestamp'] < end_date)]


# clean dtypes
aal_baggage_df['Total (Thousands)'] = aal_baggage_df['Total (Thousands)'].str.replace(",", "").astype(float)

```

```python
# clean asm

# drop the total columns in order to convert to datetime. aggregation can easily replace total
aal_asm_df = aal_asm_df.loc[aal_asm_df['Month'] != "TOTAL"].copy()

# create datetime column to index compare to stock. 
aal_asm_df['date_str'] = aal_asm_df['Year'].astype(str) + '-' + aal_asm_df['Month'].astype(str) + '-01'
aal_asm_df['timestamp'] = pd.to_datetime(aal_asm_df['date_str'])

# put time stamp as first column for clarity
aal_asm_df = aal_asm_df[['timestamp'] + [col for col in aal_asm_df.columns if col != 'timestamp']]

# since these are monthly aggregates, set date to end of month
aal_asm_df['timestamp'] = aal_asm_df['timestamp'] + pd.offsets.MonthEnd(0)

#drop placeholder date_str
aal_asm_df.drop(columns=['date_str'], inplace=True)

# filter df for dates that align with stock data
start_date = pd.to_datetime('2016-10-01')
end_date = pd.to_datetime('2018-05-01')
aal_asm_df = aal_asm_df.loc[(aal_asm_df['timestamp'] >= start_date) & (aal_asm_df['timestamp'] < end_date)]

# convert dtypes
aal_asm_df['Domestic (Thousands)'] = aal_asm_df['Domestic (Thousands)'].str.replace(",", "").astype(float)
aal_asm_df['International (Thousands)'] = aal_asm_df['International (Thousands)'].str.replace(",", "").astype(float)
aal_asm_df['Total (Thousands)'] = aal_asm_df['Total (Thousands)'].str.replace(",", "").astype(float)
```
